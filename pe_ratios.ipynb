**P/E Ratios - SP500/MSCI World - Jupyter Notebook, Python 3.12.1, September 2024**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'data/pe_ratios_clean.csv'
df = pd.read_csv(file_path)

# Convert 'Date' to datetime without specifying the format to handle multiple possible formats
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Plot the data
plt.figure(figsize=(10,6), dpi=300)

# Plot S&P 500 and MSCI World P/E Ratios
plt.plot(df['Date'], df['SP500'], label='Ratio P/E - S&P 500', marker='o')
plt.plot(df['Date'], df['MSCI World'], label='Ratio P/E - MSCI World', marker='s')

# Adding title and labels
plt.title('Ratios P/E du S&P 500 et MSCI World - 1999 à 2024')
plt.xlabel('Date')
plt.ylabel('P/E Ratio')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Adding grid and legend for better clarity
plt.grid(True)
plt.legend()

# Adjust layout for better fit
plt.tight_layout()

# Display the plot
plt.show()
```
